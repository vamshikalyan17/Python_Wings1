```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = np.loadtxt('FuelEconomy.csv', delimiter=',')
X = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.3, random_state=42)

###########################
##### Start from here #####
###########################

alg = LinearRegression()

def fit(x_train, y_train):
    ## Write code for fit function
    alg.fit(x_train, y_train)
    return alg.coef_[0], alg.intercept_
    


def predict(x, m, c):
    ## Write code for predict function
    y = m*x + c
    return alg.predict(y)
    

def score(y_truth, y_pred):
    ## Write code for score function
    return alg.score(y_truth, y_pred)
    


def cost(x, y, m, c):
    ## Write code for cost function
    x = x
    y = y
    
    # Calculate predictions
    predictions = m * x + c
    
    # Calculate errors
    errors = predictions - y
    
    # Calculate MSE
    mse = np.mean(errors ** 2)
    
    return mse


###########################
###########################

m, c = fit(X_train, Y_train)

y_test_pred = predict(X_test, m, c)
y_train_pred = predict(X_train, m, c)

print(round(score(Y_test, y_test_pred), 2))
print(round(score(Y_train, y_train_pred), 2))
print(round(cost(X_train,Y_train, m, c ), 2))
```

```python
## Open and read data file as specified in the question
## Print the required output in given format
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = np.loadtxt('FuelEconomy.csv', delimiter=',')
X = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.3, random_state=42)

###########################
##### Start from here #####
###########################

# alg = LinearRegression()

def fit(x_train, y_train):
    ## Write code for fit function
    num = (x_train*y_train).mean() - x_train.mean() * y_train.mean()
    den = (x_train**2).mean() - x_train.mean()**2
    m = num/den
    c = y_train.mean() - m * x_train.mean()
    return m, c
    


def predict(x, m, c):
    ## Write code for predict function
    y = m*x + c
    # alg.predict(y)
    return y
    

def score(y_truth, y_pred):
    ## Write code for score function
    # return alg.score(y_truth, y_pred)
    u = ((y_truth - y_pred)**2).sum()
    v = ((y_truth - y_truth.mean())**2).sum()
    return 1 - u/v
    


def cost(x, y, m, c):
    ## Write code for cost function
    # pred = m * x + c
    # errors = y - pred
    # msc = np.mean(errors**2)
    # return msc
    return ((y - m * x - c)**2).mean()


###########################
###########################

m, c = fit(X_train, Y_train)

y_test_pred = predict(X_test, m, c)
y_train_pred = predict(X_train, m, c)

print(round(score(Y_test, y_test_pred), 2))
print(round(score(Y_train, y_train_pred), 2))
print(round(cost(X_train,Y_train, m, c ), 2))
```
