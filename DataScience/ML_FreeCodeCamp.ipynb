```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
col = ["fLength", "fWidth", "fSize", "fConc", "fConc1", "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"]
df = pd.read_csv("magic04.data", names = col)
df.head()
```

```python
# Convert the class column g = 1 and h = 0
df["class"] = (df["class"] == "g").astype(int)
```

```python
df.head()
```

```python
for label in col[:-1]:
    plt.hist(df[df["class"] == 1][label], color = "blue", label = "alpha", alpha = 0.7, density = True)
    plt.hist(df[df["class"] == 0][label], color = "red", label = "gama", alpha = 0.7, density = True)
    plt.title(label)
    plt.ylabel("probablity")
    plt.xlabel(label)
    plt.legend()
    plt.show()
#     print(df[df["class"] == 1][i])
```
