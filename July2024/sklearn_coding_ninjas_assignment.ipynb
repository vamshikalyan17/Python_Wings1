```python
from sklearn import datasets
iris = datasets.load_Diabetes()
x = iris.data
y = iris.target

from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.30, train_size=0.70)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```
