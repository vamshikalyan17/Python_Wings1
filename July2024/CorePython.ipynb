'backslashreplace'	- uses a backslash instead of the character that could not be encoded
'ignore'	- ignores the characters that cannot be encoded
'namereplace'	- replaces the character with a text explaining the character
'strict'	- Default, raises an error on failure

```python
class Vamshi:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Krishna:
    def __init__(self, li):
        self.li = li
    
    def age_validation(self):
        age = []
        for i in self.li:
            if i.age >= 18:
                age.append(i)
        return age

li = []
n = int(input("Enter the value of n"))

#append the Vamshi constructor with values to li
for i in range(n):
    l = list(map(str, input().split()))
    li.append(Vamshi(l[0], int(l[1])))

obj = Krishna(li)
for i in obj.age_validation():
    print(i.name, i.age)
```

```python
class Vamshi:
    def __iter__(self):
        self.a = 1
        return self
    
    def __next__(self):
        if self.a < 8:
            x = self.a
            self.a += 1
            return x
        else:
            raise StopIteration

obj = Vamshi()
it = iter(obj)

for i in it:
    print(i)
```
