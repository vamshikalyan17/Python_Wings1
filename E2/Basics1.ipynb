# String Formatting

```python
x = "Vamshi"
y = 23
print(f"My name is {x} and my age is {y}")
```

```python
print("Hello {1}, your age is {0}".format("Vamshi", "23"))
```

```python
print("My name is {a} and my age is {b}".format(a = "Vamshi", b = "23"))
```

```python
name = input()
age = int(input())
print(f"My name is {name} and my age is {age:.3f}")
```

```python
try:
    name = input()
    age = int(input())
except Exception as e:
    print("check if you have entered the correct data type")
else:
    print(f"My name is {name} and my age is {age}")
finally:
    print("executed")
```
